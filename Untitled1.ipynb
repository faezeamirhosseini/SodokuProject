```python
import numpy as np
import pandas as pd
import mglearn
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print('keys of dataset:',iris_dataset.keys)
```

```python
print(iris_dataset['DESCR'][:193]+'\n')
```

```python
print('target names:',iris_dataset['target_names'])
```

```python
print('feature names:',iris_dataset['feature_names'])
```

```python
print('type of data:',type(iris_dataset['target']))
```

```python
print('shape of target',iris_dataset['target'].shape)
```

```python
print('target',iris_dataset['target'])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(
    iris_dataset['data'],iris_dataset['target'] , random_state=0
)
```

```python
print('x_train shape:', x_train.shape)
print('y_train shape : ' , y_train.shape)
print('x_test shape :' , x_test.shape)
print('x_test shape :' , x_test.shape)
```

```python
iris_datafram = pd.DataFrame(x_train ,columns= iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_datafram , c=y_train ,
                           figsize=(15,15),marker='x',
                           hist_kwds={'bins': 20}, s=60,
                           alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train , y_train)
```

```python
x_new = np.array([[5,2.9,1,0.2]])
print('x_new shape:' , x_new.shape)
```

```python
prediction = knn.predict(x_new)
print('prediction',prediction)
print('prediction target name:',iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(x_test)
print('test set y prediction:\n',y_pred)
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
print("Test set score : {:.2f}".format(knn.score(x_test,y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
